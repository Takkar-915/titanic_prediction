```python
#教師データとテストデータ、双方にエンジニアリングを行うため、いったん一つに結合する
data_all = pd.concat([data_train, data_test], sort = False)
data_all
```

```python
#各特徴量の欠損値の数を確認
#ここで、test_dataにはSurvivedのデータが存在しないため、結合後のデータフレームでは欠損値として扱われる

data_all.isnull().sum()
```

EDAをもとに仮説を立て、新たな特徴量を追加する

```python
#男性と比較して女性の生存率が高いため、男性と女性の別々の特徴量に分けてみる
data_all["Sex_male"] = data_all["Sex"].replace(["male", "female"],[1,0])
data_all["Sex_female"] = data_all["Sex"].replace(["male","female"], [0,1])
data_all["Sex"] = data_all["Sex"].replace(["male", "female"], [0,1])
data_all.head()
```

```python
#ParchとSibSpはどちらも家族構成の特徴量である上に、人数によって生存率が異なる傾向にあることから、2つを組み合わせた特徴量を作成する
#単身者を反映させるために+1
data_all["Families"] = data_all["Parch"] + data_all["SibSp"] + 1 
data_train["Families"] = data_all["Families"][:len(data_train)]
data_test["Families"] = data_all["Families"][len(data_train):]
sns.countplot(x = "Families",data= data_train, hue = "Survived")
```

```python
#1(単身者が多いため、別の特徴量としてわける)
data_all["Alone"] = 0
data_all.loc[data_all["Families"] == 1, "Alone"] = 1
data_all.head()
```

```python
#Ageはこれまでは平均値で欠損値を補完していたが、年代差が大きいため、平均に標準偏差を加えてばらつきを考慮した補完をする
ave = data_all["Age"].mean()
std = data_all["Age"].std()
#平均±標準偏差間の値からランダムで選出された値で欠損値を補完する
#乱数で補完しているため、再現性がない。中央値で補完してみるのも手として考えられる
data_all["Age"].fillna(np.random.randint(ave - std, ave + std), inplace = True)
data_all.head()
```

```python
#Sexの値を数値の置き換える　ここでは簡単にreplaceメソッドを用いる
#inplaceは変更の結果を元のデータフレームに反映させるか否かを指定する
#data_all["Sex"].replace(["male","female"], [0,1], inplace=True)
```

```python
#data_all.head
```

```python
#embarkedの欠損地を補完し、数値に置き換える
#本来はモデルの精度を比較して補完する必要がある
data_all["Embarked"].fillna("S",inplace = True)
data_all["Embarked"].replace(["S","C","Q"],[0, 1, 2], inplace = True)
```

```python
data_all
```

```python
#Fareの欠損値、Ageの欠損地を平均値で補完する　もともと数字なので変換の必要はない
#補完の方法として、欠損値があるデータ自体を削除したり、最頻値や中央値を用いたり、そもそも特徴量として採用しないことなどがある
data_all["Fare"].fillna(np.mean(data_all["Fare"]), inplace = True)
data_all["Age"].fillna(np.mean(data_all["Age"]), inplace = True)
```

```python
#影響が少ないと考えられる特徴量はいったん削除する
drop_columns = ["PassengerId","Name", "Ticket", "Cabin"]
#axis=1で列方向に削除する。
data_all.drop(drop_columns, axis = 1, inplace = True)
```

```python
#欠損値ないか確認
data_all.isnull().sum()
```

```python
#結合したデータを教師データとテストデータに分離
#先頭から教師データ行数分のデータを抜き出している
data_train = data_all[:len(data_train)]
#教師データの行数分の１個下のデータから最後まで
data_test = data_all[len(data_train):]
```

モデル作成

```python
#多くのアルゴリズムは特徴量と目的変数を分離して投入
y_data_train = data_train["Survived"]
x_data_train = data_train.drop("Survived", axis = 1)
x_data_test = data_test.drop("Survived", axis = 1)
```

```python
#ランダムフォレストアルゴリズムをインポート
from sklearn.ensemble import RandomForestClassifier

#引数はそれぞれ決定木の数,木の深さ(階層),乱数
clf = RandomForestClassifier(n_estimators = 100, max_depth = 2, random_state = 0)
```

```python
#教師データの学習
clf.fit(x_data_train, y_data_train)
```

```python
#作成したモデルで予測
#predictの閾値は0.5が閾値なので,それを基準に0,1に変換
y_data_pred = clf.predict(x_data_test)
```

```python
y_data_pred
```

submitの作業
kaggleで用意されている提出用のフォーマットに従う必要あり

```python
#予測結果をsubmit用のcsvとして作成
submit = data_gender_submission
submit["Survived"] = list(map(int, y_data_pred))
submit.to_csv("new_engineering_submit.csv", index = False)
#indexは保存しないという意
```
